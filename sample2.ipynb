```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./models/Delhi.csv')
df
```

```python
# df['date'] = pd.to_datetime(df['last_update']).dt.date
# df['time'] = pd.to_datetime(df['last_update']).dt.time
df_filtered = df[df['city'] == 'Delhi']
df_filtered

```

```python
df_grouped = df_filtered.groupby('station')[['pollutant_min','pollutant_max','pollutant_avg']].mean()

df_grouped.plot(kind='bar', figsize=(14,6))
plt.xlabel('Station')
plt.ylabel('Pollutant Levels')
plt.title('Pollutant Levels per Station (Delhi)')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

df_grouped = df_filtered.groupby('station')['pollutant_avg'].mean()

plt.figure(figsize=(12,6))
df_grouped.sort_values().plot(kind='bar')

plt.xlabel('Station')
plt.ylabel('Average Pollutant Level')
plt.title('Average Pollutant Level per Station (Delhi)')
plt.tight_layout()
plt.show()
```
